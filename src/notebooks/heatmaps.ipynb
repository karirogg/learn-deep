```python
import matplotlib.pyplot as plt
import pickle
import torch
```

```python
n = 2
```

```python
with open(f'../../checkpoints/n_{n}_epochs_30_epoch_class_mat.pt', 'rb') as f:
    epoch_class_mat = pickle.load(f)
```

```python
for i, task in enumerate(range(n)):
# plot heatmap of classification accuracy per sample
    order = torch.argsort(torch.mean(epoch_class_mat[i][:,i,:], axis=1), descending=False)

    concat_task_progression = torch.cat([epoch_class_mat[i][:, j, :] for j in range(n)], dim=1)[order]

    plt.figure(figsize=(5 * n, 5))
    plt.imshow(concat_task_progression.cpu().numpy(), cmap='cividis', interpolation='nearest', aspect='auto')

    plt.show()
```
